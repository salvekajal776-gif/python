```python
# File handling in Python is the process of creating, reading, writing, updating, 
# and deleting files using functions like open() 
# and file modes.
```

```python
# Create a Sample File
```

```python
with open("sample.txt", "w") as f:
    f.write("Name: Rahul\n")
    f.write("Role: Data Scientist\n")
    f.write("Location: India\n")
```

```python
# Reading the File
```

```python
with open("sample.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
# Reading Line by Line
```

```python
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
# Writing (Overwrite the File)
```

```python
with open("sample.txt", "w") as f:
    f.write("This is new content\n")
```

```python
# Appending Data
```

```python
with open("sample.txt", "a") as f:
    f.write("Appended line 1\n")
    f.write("Appended line 2\n")
```

```python
# Read Again After Append
```

```python
with open("sample.txt", "r") as f:
    print(f.read())
```

```python
# Check if File Exists
```

```python
import os

if os.path.exists("sample.txt"):
    print("File exists")
else:
    print("File not found")
```

```python
# Delete the File
```

```python
import os
os.remove("sample.txt")
```
