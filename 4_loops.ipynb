```python
# Loops
```

```python
# A for loop in Python repeats a block of code a specific number of times by stepping through a collection of items.
```

```python
# Print numbers from 0 to 4
```

```python
for i in range(5):
    print(i)
```

```python
# Print numbers from 1 to 5
```

```python
for i in range(1, 6):
    print(i)
```

```python
fruits = ["Apple", "Banana", "Mango"]
```

```python
for fruit in fruits:
    print(fruit)
```

```python
# Multiplication table
```

```python
num = 5
```

```python
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# while Loop
```

```python
# A while loop in Python runs the same block of code over and over again as long as a specific condition is True.
```

```python
# numbers from 1 to 5
```

```python
i = 1
```

```python
while i <= 5:
    print(i)
    i += 1
```

```python
# Reverse counting
```

```python
i = 5
```

```python
while i >= 1:
    print(i)
    i -= 1
```

```python
# Password checking
```

```python
password = ""
```

```python
while password != "python":
    password = input("Enter Password: ")

print("Login Successful")
```

```python
# break
```

```python
# The break statement in Python instantly stops and exits a loop, even if the loop's condition is still True
```

```python
# Stop loop at 5
```

```python
for i in range(1, 11):

    if i == 5:
        break

    print(i)
```

```python
# ATM exit program
```

```python
while True:

    amount = int(input("Enter Amount (0 to Exit): "))

    if amount == 0:
        break

    print("Amount Entered:", amount)
```

```python
# continue
```

```python
# The continue statement in Python skips the rest of the code inside the current loop iteration and jumps straight to the next turn
```

```python
# Skip number 5
```

```python
for i in range(1, 11):

    if i == 5:
        continue

    print(i)
```

```python
# Print only odd numbers
```

```python
for i in range(1, 21):

    if i % 2 == 0:
        continue

    print(i)
```

```python
# Nested Loops
```

```python
# Print coordinates
```

```python
for i in range(1, 4):

    for j in range(1, 4):

        print(i, j)
```

```python
# Square star pattern
```

```python
for i in range(5):

    for j in range(5):

        print("*", end=" ")

    print()
```
